### Vamos ver como Podemos melhorar plots usando matplotlib

Vamos imaginar que temos a quantidade de vendas mensais de dois produtos A e B de uma empresa, e queremos fazer uma visualização gráfica
desses dados


```python
import matplotlib.pyplot as plt
import numpy as np


# Definição dos dados
x_data = np.arange(0, 10)
y1_data = np.array([160, 200, 180, 210 , 220 , 300 , 280 , 310 , 270 , 250 , 260, 220 ]) # Produto A
y2_data = [170, 210, 180, 215 , 220 , 340 , 310 , 320 , 325 , 342 , 359, 342 ]           # Produto B

meses = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 5))

# Plot dos dados
ax.plot(meses, y1_data, label = 'Produto A')
ax.plot(meses, y2_data, label  = 'Produto B ')

ax.legend(loc='upper left')


```

### Vamos retirar as bordas superior e direita
### Setar y limite 


```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 5))

# Plot dos dados
ax.plot(meses, y1_data, label = 'Produto A')
ax.plot(meses, y2_data, label  = 'Produto B')

ax.legend(loc='upper left')


# Retirando as bordas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

## ajustar o y_lim

# Ajustando y_lim
ax.set_ylim(100, 440)
```

### Arrumando o eixo x ... 

Estamos percebendo que representar os meses por extenso é uma péssima forma de representação

```python
# Definição dos dados
x_data = np.arange(0, 10)
y1_data = np.array([160, 200, 180, 210 , 220 , 300 , 280 , 310 , 270 , 250 , 260, 220 ]) # Produto A
y2_data = [170, 210, 180, 215 , 220 , 340 , 310 , 320 , 325 , 342 , 359, 342 ]           # Produto B

meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 5))

# Plot dos dados
ax.plot(meses, y1_data, label = 'Produto A')
ax.plot(meses, y2_data, label  = 'Produto B')

ax.legend(loc='upper left')


# Retirando as bordas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

## ajustar o y_lim

# Ajustando y_lim
ax.set_ylim(100, 440)
```

## Setar título e descrição no eixo y


```python
meses = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 5))

# Plot dos dados
ax.plot(meses, y1_data, label = 'Produto A', color = 'gray')
ax.plot(meses, y2_data, label  = 'Produto B' , color = 'blue')

ax.legend(loc='upper left')


# Retirando as bordas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

## ajustar o y_lim

# Ajustando y_lim
ax.set_ylim(100, 440)

## setar título e descrição

# Colocando descrição no eixo y
ax.set_ylabel('Número de Vendas', fontsize=12)
ax.yaxis.set_label_coords(-0.08, 0.75)
ax.text(-1, 460, 'Comparação de vendas entre Produtos ao longo do tempo', fontsize=13, fontweight='bold')
```

### Bolinhas nos últimos 5 dados 
### Enfatizando valores nas ultimas marcações

```python


# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 5))

# Plot dos dados
ax.plot(meses, y1_data, label = 'Produto A', color = 'gray')
ax.plot(meses, y2_data, label  = 'Produto B' , color = 'blue')

ax.legend(loc='upper left')


# Retirando as bordas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

## ajustar o y_lim

# Ajustando y_lim
ax.set_ylim(100, 440)

## setar título e descrição

# Colocando descrição no eixo y
ax.set_ylabel('Número de Vendas', fontsize=12)
ax.yaxis.set_label_coords(-0.08, 0.75)
ax.text(-1, 460, 'Comparação de vendas entre Produtos ao longo do tempo', fontsize=13, fontweight='bold')

# Colocando bolinhas nos últimos 5 marcadores
ax.scatter(meses[-5:], y1_data[-5:], color='gray')
ax.scatter(meses[-5:], y2_data[-5:], color='blue')

## colocar valores nas ultimas marcações

for i in range(5):

    ax.annotate(f'{y1_data[11-i]}' ,(meses[11-i] , y1_data[11-i]),
                xytext = (+1, -17) , textcoords = 'offset points' , color= 'gray',
                ha = 'center')
    
    ax.annotate(f'{y2_data[11-i]}' ,(meses[11-i] , y2_data[11-i]),
            xytext = (+1, +12) , textcoords = 'offset points' , color= 'blue',
            ha = 'center')
```

## Texto descritivo sobre mudança de estratégia:  



```python
# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 5))

# Plot dos dados
ax.plot(meses, y1_data, label = 'Produto A', color = 'gray')
ax.plot(meses, y2_data, label  = 'Produto B' , color = 'blue')

ax.legend(loc='upper left')


# Retirando as bordas
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

## ajustar o y_lim

# Ajustando y_lim
ax.set_ylim(100, 440)

## setar título e descrição

# Colocando descrição no eixo y
ax.set_ylabel('Número de Vendas', fontsize=12)
ax.yaxis.set_label_coords(-0.08, 0.75)
ax.text(-1, 460, 'Comparação de vendas entre Produtos ao longo do tempo', fontsize=13, fontweight='bold')

# Colocando bolinhas nos últimos 5 marcadores
ax.scatter(meses[-5:], y1_data[-5:], color='gray')
ax.scatter(meses[-5:], y2_data[-5:], color='blue')

## colocar valores nas ultimas marcações

for i in range(5):

    ax.annotate(f'{y1_data[11-i]}' ,(meses[11-i] , y1_data[11-i]),
                xytext = (+1, -17) , textcoords = 'offset points' , color= 'gray',
                ha = 'center')
    
    ax.annotate(f'{y2_data[11-i]}' ,(meses[11-i] , y2_data[11-i]),
            xytext = (+1, +12) , textcoords = 'offset points' , color= 'blue',
            ha = 'center')

## adicionando um comentario pertinente ao gráfico com viés de explicação
ax.axvline(x = 4, color='gray' , linestyle = '--')
texto_1 = 'Mudança de estratégia de Marketing'
texto_2 = '''Essa mudança realizada em maio foi eficaz com o produto B, 
O Produto A não manteve o crescimento '''

ax.text(4 + 0.1  , ax.get_ylim()[1]*0.98 , texto_1, fontsize = 10 , color= 'gray',
        fontweight= 'bold')

ax.text(4 + 0.1  , ax.get_ylim()[1]*0.9 , texto_2, fontsize = 10 , color= 'gray')
plt.savefig('Vendas.jpeg' , dpi = 600)
plt.show()
```
